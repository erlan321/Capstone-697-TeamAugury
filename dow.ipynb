```python
one_hot_dow_hod = X_transf_df.iloc[:,4:35]
one_hot_dow_hod.shape
```

```python
cor_data_simple = one_hot_dow_hod.corr().stack().reset_index().rename(columns={0: 'correlation', 'level_0': 'variable', 'level_1': 'variable2'})
cor_data_simple['correlation_label'] = cor_data_simple['correlation'].map('{:.2f}'.format)
cor_data_simple.head()
```

```python
cor_data_simple.variable2.unique()
```

```python
base = alt.Chart(cor_data_simple).encode(
    x='variable2:O',
    y='variable:O'    
)

# Text layer with correlation labels
# Colors are for easier readability
text = base.mark_text().encode(
    text='correlation_label',
    color=alt.condition(
        alt.datum.correlation > 0.5, 
        alt.value('white'),
        alt.value('black')
    )
)

# The correlation heatmap itself
cor_plot = base.mark_rect().encode(
    color='correlation:Q'
).properties(height=800, width = 800)

cor_plot + text # The '+' means overlaying the text and rect layer
```

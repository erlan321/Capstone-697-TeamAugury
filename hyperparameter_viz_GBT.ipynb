# Holding "all else equal" let's look at the median result for each parameter option to identify any trends...

```python
df = csv_data.copy()
df = df[df.Score=='F1']
#print(df.columns)
df = df.groupby(['type','learning_rate',])['Result'].median().reset_index()

chart1 = alt.Chart(df).mark_line().encode(
    x = 'learning_rate:O',
    y = 'Result:Q',
    color = 'type:N',
    tooltip = 'learning_rate:Q',

).properties(
    title = 'F1 Score'
)

df = csv_data.copy()
df = df[df.Score=='F1']
df = df.groupby(['type','n_estimators',])['Result'].median().reset_index()


chart2 = alt.Chart(df).mark_line().encode(
    x = 'n_estimators:O',
    y = 'Result:Q',
    color = 'type:N',
    tooltip = 'n_estimators:Q',

).properties(
    title = 'F1 Score'
)

df = csv_data.copy()
df = df[df.Score=='F1']
#print(df.columns)
df = df.groupby(['type','max_depth',])['Result'].median().reset_index()
#display(df)

chart3 = alt.Chart(df).mark_line().encode(
    x = 'max_depth:O',
    y = 'Result:Q',
    color = 'type:N',
    tooltip = 'max_depth:Q',

).properties(
    title = 'F1 Score'
)

df = csv_data.copy()
df = df[df.Score=='F1']
#print(df.columns)
df = df.groupby(['type','max_features',])['Result'].median().reset_index()

chart4 = alt.Chart(df).mark_line().encode(
    x = 'max_features:O',
    y = 'Result:Q',
    color = 'type:N',
    tooltip = 'max_features:Q',

).properties(
    title = 'F1 Score'
)

df = csv_data.copy()
df = df[df.Score=='F1']
#print(df.columns)
df = df.groupby(['type','subsample',])['Result'].median().reset_index()

chart5 = alt.Chart(df).mark_line().encode(
    x = 'subsample:O',
    y = 'Result:Q',
    color = 'type:N',
    tooltip = 'subsample:Q',

).properties(
    title = 'F1 Score'
)

chart1 | chart2 | chart3 | chart4 | chart5
```

```python
df = csv_data.copy()
df = df[df.Score=='Accuracy']
#print(df.columns)
df = df.groupby(['type','learning_rate',])['Result'].median().reset_index()

chart1 = alt.Chart(df).mark_line().encode(
    x = 'learning_rate:O',
    y = 'Result:Q',
    color = 'type:N',
    tooltip = 'learning_rate:Q',

).properties(
    title = 'Accuracy'
)

df = csv_data.copy()
df = df[df.Score=='Accuracy']
df = df.groupby(['type','n_estimators',])['Result'].median().reset_index()


chart2 = alt.Chart(df).mark_line().encode(
    x = 'n_estimators:O',
    y = 'Result:Q',
    color = 'type:N',
    tooltip = 'n_estimators:Q',

).properties(
    title = 'Accuracy'
)

df = csv_data.copy()
df = df[df.Score=='Accuracy']
#print(df.columns)
df = df.groupby(['type','max_depth',])['Result'].median().reset_index()
#display(df)

chart3 = alt.Chart(df).mark_line().encode(
    x = 'max_depth:O',
    y = 'Result:Q',
    color = 'type:N',
    tooltip = 'max_depth:Q',

).properties(
    title = 'Accuracy'
)

df = csv_data.copy()
df = df[df.Score=='Accuracy']
#print(df.columns)
df = df.groupby(['type','max_features',])['Result'].median().reset_index()

chart4 = alt.Chart(df).mark_line().encode(
    x = 'max_features:O',
    y = 'Result:Q',
    color = 'type:N',
    tooltip = 'max_features:Q',

).properties(
    title = 'Accuracy'
)

df = csv_data.copy()
df = df[df.Score=='Accuracy']
#print(df.columns)
df = df.groupby(['type','subsample',])['Result'].median().reset_index()

chart5 = alt.Chart(df).mark_line().encode(
    x = 'subsample:O',
    y = 'Result:Q',
    color = 'type:N',
    tooltip = 'subsample:Q',

).properties(
    title = 'Accuracy'
)

chart1 | chart2 | chart3 | chart4 | chart5
```
